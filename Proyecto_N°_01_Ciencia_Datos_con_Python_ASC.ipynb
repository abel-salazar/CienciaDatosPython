##**Ejercicio N°1:**
Escribir un programa que pregunte al usuario o genere las ventas
de los últimos 5 días, y muestre por pantalla una serie (Dataframe)
con los datos de las ventas por producto (opcional): Nro de
unidades y precio unitario, indexada por los días, antes y después
de aplicarles un descuento del 5%.
(Diversas opciones de solución: libre)

```python
# Vamos importar la libreria pandas. Vamos a utilizar esta libreria en todos los ejercicios, por lo que lo vamos a
# importar sólo 1 vez.
import pandas as pd
```

```python
# Solicitamos al usuario que ingrese las ventas por producto de los últimos 5 días
# Inicializamos un array para almacenar los datos
ventas_por_producto = []
# Realizamos un bucle for para 5 ingresos, solictando: Unidades Vendidas y Precio Unitario
for i in range(5):
    unidades = int(input(f"Ingrese el número de unidades vendidas el día {i+1}: "))
    precio_unitario = float(input(f"Ingrese el precio unitario del producto vendido el día {i+1}: "))
# Utilizamo append para poblar la matriz
    ventas_por_producto.append({'Día': f'Día {i+1}', 'Unidades': unidades, 'Precio Unitario': precio_unitario})
```

```python
# Creamos un DataFrame con los datos de ventas por producto
df_ventas_por_producto = pd.DataFrame(ventas_por_producto)
```

```python
# Calculamos el total de ventas por producto
df_ventas_por_producto['Total'] = df_ventas_por_producto['Unidades'] * df_ventas_por_producto['Precio Unitario']
```

```python
# Aplicamos un descuento del 5%
df_ventas_por_producto['Total con Descuento'] = df_ventas_por_producto['Total'] * 0.95
```

```python
# Mostramos el DataFrame con los datos de las ventas por producto antes y después del descuento
print("DataFrame con los datos de las ventas por producto:")
print(df_ventas_por_producto)
```

##**Ejercicio Nº 2**
Escribir un programa que reciba un diccionario con los ingresos
mensuales de los clientes de una empresa, y devuelva una serie
con el ingreso mínimo, máximo y media.
El resultado final debe ser una serie con los estadísticos.

```python
# Primero, vamos a definir una función llamada calcular_estadisticas que tomara un diccionario de ingresos mensuales y
# devolverá una serie con el ingreso mínimo, máximo y medio.
# Utilizamo la sentencia: def para definir nuestra función.
def calcular_estadisticas(ingresos):
    # Convertir el diccionario de ingresos a una serie
    serie_ingresos = pd.Series(ingresos)

    # Calcular el ingreso mínimo, máximo y media
    ingreso_minimo = serie_ingresos.min()
    ingreso_maximo = serie_ingresos.max()
    ingreso_media = serie_ingresos.mean()

    # Crear una serie con los estadísticos
    estadisticos = pd.Series([ingreso_minimo, ingreso_maximo, ingreso_media], index=['Mínimo', 'Máximo', 'Media'])

    return estadisticos
```

```python
# Definimos el diccionario de ingresos mensuales
ingresos_mensuales = {'Enero': 5000, 'Febrero': 6000, 'Marzo': 7000, 'Abril': 8000, 'Mayo': 9000}
```

```python
# Llamamos a la función calcular_estadisticas con el diccionario de ingresos
estadisticas_ingresos = calcular_estadisticas(ingresos_mensuales)
```

```python
# Mostramos la serie con los estadísticos
print("Estadísticas de ingresos:")
print(estadisticas_ingresos)
```

##**Ejercicio Nº 3**
Escribir programa que genere y muestre por pantalla un DataFrame
con los datos de la siguiente tabla:

Mes     Ventas    Gastos
Enero   30500     22000
Febrero 35600     23450
Marzo   28300     18100
Abril   33900     35700
Mayo    42500     32450

A continuación, generar nuevas columna:
1. Columna 1 = Gasto/Ventas*100.
2. Columna 2 = Si Ventas>= 30000, imprimir “Meta Superada”,
caso contrario imprimir “Meta no Superada”.
3. Columna 3 = Si Gasto>Ventas, calcular el sobre gasto.
4. Obtener el total de las ventas de los últimos 3 meses
5. Obtener el total de gastos de febrero a abril

**SOLUCIÓN:**
A continuación, vamos a crear un DataFrame con los datos proporcionados, calculando las nuevas columnas requeridas, y obteniendo los totales de ventas de los últimos 3 meses y los gastos de febrero a abril. Luego, podremos mostrar el DataFrame resultante junto con los totales calculados.

```python
# Creamos un DataFrame con los datos proporcionados
datos = {
    'Mes': ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo'],
    'Ventas': [30500, 35600, 28300, 33900, 42500],
    'Gastos': [22000, 23450, 18100, 35700, 32450]
}

df_datos = pd.DataFrame(datos)
```

```python
# Calculamos la columna de gasto por ventas y agregarla al DataFrame
df_datos['Gasto/Ventas (%)'] = (df_datos['Gastos'] / df_datos['Ventas']) * 100
```

```python
# Calcular si se superó la meta de ventas y agregarlo como una nueva columna
df_datos['Meta Superada'] = ['Meta Superada' if ventas >= 30000 else 'Meta no Superada' for ventas in df_datos['Ventas']]
```

```python
# Calculamos el sobregasto y agregarlo como una nueva columna
df_datos['Sobregasto'] = df_datos.apply(lambda row: row['Gastos'] - row['Ventas'] if row['Gastos'] > row['Ventas'] else 0, axis=1)
```

```python
# Obtenemos el total de las ventas de los últimos 3 meses
total_ventas_ultimos_3_meses = df_datos.tail(3)['Ventas'].sum()
```

```python
# Obtenemos el total de gastos de febrero a abril
total_gastos_feb_abr = df_datos[1:4]['Gastos'].sum()
```

```python
# Mostraremos el DataFrame resultante y los totales
print("DataFrame con los datos y las nuevas columnas:")
print(df_datos)

print("\nTotal de ventas de los últimos 3 meses:", total_ventas_ultimos_3_meses)
print("Total de gastos de febrero a abril:", total_gastos_feb_abr)
```
